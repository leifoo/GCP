### Definition of cloud computing
A way of using I.T. that has 5 equally important traits:
- On-demand self-service
- Broad network access
- Resource pooling
- Rapid elasticity
- Measured services

### GCP computing architectures
Virtualized data centers brought you Infrastructure as a Service, IaaS, and Platform as a Service, PaaS offerings. IaaS offerings provide raw compute, storage, and network organized in ways that are familiar from data centers. PaaS offerings, on the other hand, bind application code you write to libraries that give access to the infrastructure your application needs. In the IaaS model, you pay for what you allocate. In the PaaS model, you pay for what you use. 

### GCP regions and zones
multi-regions, region, zone

Spreading GCP resources across multiple zones in a region is protection against failure.

## Module Introduction
When you run your workloads in GCP, you use projects to organize them. You use Google Cloud Identity, and Access Management, also called IM, or IAM to control who can do what? And you use your choice of several interfaces to connect. In this module, you'll use these basics to get started. Projects are the main way you organize the resources you use in GCP. The principle of least privilege is very important in managing any kind of compute infrastructure, whether it's in the Cloud or on-premises. This principle says that each user should have only those privileges needed to do their jobs. In a least privilege environment, people are protected from an entire class of errors. GCP customers use IM to implement least privilege. 4 ways to interact with GCP's management layer. Through the web-based console, through the SDK, and its command-line tools, through the APIs, and through a mobile app. 

## GCP resource hierarchy
Organization, folder, project

Folders require an organization node. Organization nodes are optional, but if you want to create folders, having one is mandatory.

Services and APIs are enabled on a per-project basis.

## Identity and Acess Management (IAM)
3 types of IAM roles
- Primitive (offer fixed, course-grained level of access)
- Predefined (apply to a particular GCP service in a project)
- Custom (define a precise set of permissions, only on a project or organization level, not folder level)

Compute Engines InstantAdmin Role lets whoever has that role perform a certain set of actions on virtual machines. The actions are listing them, reading and changing their configurations, and starting and stopping them.

Service Account control server-to-server interactions (identity, authenticate, control privileges).

A service account is also a resource, so it can have IAM policies on its own attached to it.

### Demonstration: Getting started with Cloud Launcher
Use Cloud Launcher to deploy LAMP stack (Linux Apache MySQL, PHP) on GCP. It's an easy environment for developing web applications.

__Console__, click __left top__, click __Cloud Launcher__, type __LAMP__ in search bar (LAMP Certified by Bitnami), click __Launch On Compute Engine__, click __Deploy__
- Visit temporary home page, click __Site address__
- Configuration: 1) loging using SSH, click __SSH__; 2) change to the directory where the software is installed, `cd /opt/bitnami`; 3) copy a test page for PHP, `sudo cp docs/phpinfo.php apache2/htdocs`; 4) end SSH session, `exit`; 5) confirm PHP test page visible, __ip address/phpinfo.php__.
